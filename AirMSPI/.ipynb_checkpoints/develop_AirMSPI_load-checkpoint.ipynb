```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
import shdom
from mpl_toolkits.mplot3d import Axes3D

data_dir = '../AirMSPI/AirMSPI_ER2_GRP_ELLIPSOID_20130206_202416Z_Pacific-32N123W'
```

```python
shdom.AirMSPIMeasurementsv3().imshow(data_dir)
```

```python
%matplotlib notebook 

bs = shdom.bbox_select(data_dir,3)
```

```python
roi = []
for i in range(9):
    roi.append(shdom.AirMSPIMeasurementsv3().select_region_of_interest(data_dir,i))
```

```python
# roi = [1360, 1596, 1520, 1760]
# roi = [1752, 1928, 1468, 1696]
# roi = [1292-300, 1500, 1140-300, 1352]
# roi = [2388, 2696, 1020, 1316]



# roi = [2112, 2264, 1464, 1620]
# roi = [1232, 1428, 1100, 1320]
airmspi_measurements = shdom.AirMSPIMeasurementsv3()
airmspi_measurements.load_from_hdf(data_dir,region_of_interest=roi,valid_wavelength=[865])
```

```python
output_dir = '../Develop_Dynamic_cloud/experiments/AirMSPI/dynamic_medium_polychromatic'

airmspi_measurements.save_airmspi_measurements(output_dir)
```

```python
%matplotlib inline
f, axarr = plt.subplots(1, len(airmspi_measurements.images), figsize=(20, 20))
if isinstance(axarr,plt.Axes):
    axarr = [axarr]
for ax, image in zip(axarr, airmspi_measurements.images):
    if len(image.shape)>2 and image.shape[2]==2:
        image = np.concatenate((np.zeros((200,200,1)),image),2)
    # image -= image.min()
    # ax.imshow(image/image.max())
    ax.imshow(image)
```

```python
%matplotlib inline
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.view_init()
x_min = airmspi_measurements.camera.projection.x.min()
x_max = airmspi_measurements.camera.projection.x.max()
y_min = airmspi_measurements.camera.projection.y.min()
y_max = airmspi_measurements.camera.projection.y.max()
airmspi_measurements.plot(ax,[x_min, x_max],[y_min, y_max],[0, 20],20)
```

```python
nx = ny = 40
x_min = airmspi_measurements.camera.projection.x.min()
x_max = airmspi_measurements.camera.projection.x.max()
dx = np.round((x_max-x_min)/nx,10)
x = np.arange(x_min,x_max,dx)

y_min = airmspi_measurements.camera.projection.y.min()
y_max = airmspi_measurements.camera.projection.y.max()
dy = np.round((y_max-y_min)/ny,10)
y = np.arange(y_min,y_max,dy)

grid = shdom.Grid(x=x,
                  y=y,z=np.linspace(0, 20, 20))

measurements = airmspi_measurements
dynamic_carver = shdom.DynamicSpaceCarver(airmspi_measurements)
mask_list, dynamic_grid, cloud_velocity = dynamic_carver.carve(grid, agreement=0.5,
                    time_list = measurements.time_list, thresholds=0.02,
                    vx_max = 50, vy_max=50, gt_velocity = [0]*3)
show_mask=1
if show_mask:
    a = (mask_list[0].data).astype(int)
    shdom.cloud_plot(a)
    print(sum(sum(sum(a))))
```
